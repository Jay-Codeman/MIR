### STFT spectrum

```python
plt.figure(figsize = (12,3))


# flute

y, sr = librosa.load(os.path.join(general_path,'flute.wav'))
audio_file, _ = librosa.effects.trim(y)
S = np.abs(librosa.stft(audio_file))

img = librosa.display.specshow(librosa.amplitude_to_db(S,
                                                       ref=np.max),
                               y_axis='log', x_axis='time', ax=plt.subplot(1,4,1))

plt.title("flute", fontsize = 15);



# piano

y, sr = librosa.load(os.path.join(general_path,'piano.wav'))
audio_file, _ = librosa.effects.trim(y)
S = np.abs(librosa.stft(audio_file))
img = librosa.display.specshow(librosa.amplitude_to_db(S,
                                                       ref=np.max),
                               y_axis='log', x_axis='time', ax=plt.subplot(1,4,2))

plt.title("piano", fontsize = 15);



# tenor saxophone
y, sr = librosa.load(os.path.join(general_path,'saxophone.wav'))
audio_file, _ = librosa.effects.trim(y)
S = np.abs(librosa.stft(audio_file))
img = librosa.display.specshow(librosa.amplitude_to_db(S,
                                                       ref=np.max),
                               y_axis='log', x_axis='time', ax=plt.subplot(1,4,3))

plt.title("tensor saxphone", fontsize = 15);




# violin
y, sr = librosa.load(os.path.join(general_path,'violin.wav'))
audio_file, _ = librosa.effects.trim(y)
S = np.abs(librosa.stft(audio_file))
img = librosa.display.specshow(librosa.amplitude_to_db(S,
                                                       ref=np.max),
                               y_axis='log', x_axis='time', ax=plt.subplot(1,4,4))

plt.title("violin", fontsize = 15);
plt.tight_layout()
plt.savefig("STFT for 4 instruments.jpg")
```

#### Fast Fourier Transform

```python
plt.figure(figsize = (16, 8))


# clarinet
plt.subplot(2,4,1)

y, sr = librosa.load(os.path.join(general_path,'clarinet.wav'))
audio_file, _ = librosa.effects.trim(y)
# Default FFT window size
n_fft = 2048 # Fast Fourier Transform window size
hop_length = 512 # Frame number audio between STFT columns

# 短时傅里叶变换(STFT)
D = np.abs(librosa.stft(audio_file, n_fft = n_fft, hop_length = hop_length))

print('Shape of D object:', np.shape(D))
plt.plot(D);
plt.title("STFT in clarinet", fontsize = 15);




# distorted electric guitar

plt.subplot(2,4,2)

y, sr = librosa.load(os.path.join(general_path,'1电吉他.wav'))
audio_file, _ = librosa.effects.trim(y)

n_fft = 2048
hop_length = 512 

D = np.abs(librosa.stft(audio_file, n_fft = n_fft, hop_length = hop_length))

print('Shape of D object:', np.shape(D))
plt.plot(D);
plt.title("STFT in distorted electric guitar", fontsize = 15);


# female singer
plt.subplot(2,4,3)

y, sr = librosa.load(os.path.join(general_path,'2女歌手.wav'))
audio_file, _ = librosa.effects.trim(y)

n_fft = 2048 
hop_length = 512 
D = np.abs(librosa.stft(audio_file, n_fft = n_fft, hop_length = hop_length))

print('Shape of D object:', np.shape(D))
plt.plot(D);
plt.title("STFT in female singer", fontsize = 15);


# flute
plt.subplot(2,4,4)

y, sr = librosa.load(os.path.join(general_path,'flute.wav'))
audio_file, _ = librosa.effects.trim(y)

n_fft = 2048 
hop_length = 512 
D = np.abs(librosa.stft(audio_file, n_fft = n_fft, hop_length = hop_length))

print('Shape of D object:', np.shape(D))
plt.plot(D);
plt.title("STFT in flute", fontsize = 15);


# piano
plt.subplot(2,4,5)

y, sr = librosa.load(os.path.join(general_path,'piano.wav'))
audio_file, _ = librosa.effects.trim(y)

n_fft = 2048 
hop_length = 512 

D = np.abs(librosa.stft(audio_file, n_fft = n_fft, hop_length = hop_length))
print('Shape of D object:', np.shape(D))
plt.plot(D);
plt.title("STFT in piano", fontsize = 15);


# tenor saxophone
plt.subplot(2,4,6)

y, sr = librosa.load(os.path.join(general_path,'saxophone.wav'))
audio_file, _ = librosa.effects.trim(y)
n_fft = 2048 
hop_length = 512

D = np.abs(librosa.stft(audio_file, n_fft = n_fft, hop_length = hop_length))
print('Shape of D object:', np.shape(D))
plt.plot(D);
plt.title("STFT in tenor saxophone", fontsize = 15);


# trumpet
plt.subplot(2,4,7)

y, sr = librosa.load(os.path.join(general_path,'trumpet.wav'))
audio_file, _ = librosa.effects.trim(y)
n_fft = 2048 
hop_length = 512 
D = np.abs(librosa.stft(audio_file, n_fft = n_fft, hop_length = hop_length))

print('Shape of D object:', np.shape(D))
plt.plot(D);
plt.title("STFT in trumpet", fontsize = 15);




# violin
plt.subplot(2,4,8)

y, sr = librosa.load(os.path.join(general_path,'violin.wav'))
audio_file, _ = librosa.effects.trim(y)
n_fft = 2048 
hop_length = 512 

D = np.abs(librosa.stft(audio_file, n_fft = n_fft, hop_length = hop_length))

print('Shape of D object:', np.shape(D))
plt.plot(D);
plt.title("STFT in violin", fontsize = 15);
plt.tight_layout()
plt.savefig("STFT for 8 instruments.jpg")
```

###  spectral centroid

```python
plt.figure(figsize = (22, 10))


# clarinet
plt.subplot(2,4,1)

y, sr = librosa.load(os.path.join(general_path,'clarinet.wav'))
audio_file, _ = librosa.effects.trim(y)
# Calculate spectral centroid
spectral_centroids = librosa.feature.spectral_centroid(audio_file, sr=sr)[0]

# Calculate the time variable for visualization
frames = range(len(spectral_centroids))

# Convert frames to time(second)
t = librosa.frames_to_time(frames)

# function to normalize sound data
def normalize(x, axis=0):
    return sklearn.preprocessing.minmax_scale(x, axis=axis)

# Plot spectral centroids along a waveform
librosa.display.waveshow(audio_file, sr=sr, alpha=0.4);
plt.plot(t, normalize(spectral_centroids), color='r');
plt.title("Spectrum centroid of clarinet", fontsize = 15);



# distorted electric guitar
plt.subplot(2,4,2)

y, sr = librosa.load(os.path.join(general_path,'guitar.wav'))
audio_file, _ = librosa.effects.trim(y)
spectral_centroids = librosa.feature.spectral_centroid(audio_file, sr=sr)[0]
frames = range(len(spectral_centroids))
t = librosa.frames_to_time(frames)
def normalize(x, axis=0):
    return sklearn.preprocessing.minmax_scale(x, axis=axis)
librosa.display.waveshow(audio_file, sr=sr, alpha=0.4);
plt.plot(t, normalize(spectral_centroids), color='r');
plt.title("Spectrum centroid of distorted electric guitar", fontsize = 15);


# female singer
plt.subplot(2,4,3)
y, sr = librosa.load(os.path.join(general_path,'singer.wav'))
audio_file, _ = librosa.effects.trim(y)
spectral_centroids = librosa.feature.spectral_centroid(audio_file, sr=sr)[0]
frames = range(len(spectral_centroids))
t = librosa.frames_to_time(frames)
def normalize(x, axis=0):
    return sklearn.preprocessing.minmax_scale(x, axis=axis)
librosa.display.waveshow(audio_file, sr=sr, alpha=0.4);
plt.plot(t, normalize(spectral_centroids), color='r');
plt.title("Spectrum centroid of female singer", fontsize = 15);


# flute
plt.subplot(2,4,4)
y, sr = librosa.load(os.path.join(general_path,'flute.wav'))
audio_file, _ = librosa.effects.trim(y)
spectral_centroids = librosa.feature.spectral_centroid(audio_file, sr=sr)[0]
frames = range(len(spectral_centroids))
t = librosa.frames_to_time(frames)
def normalize(x, axis=0):
    return sklearn.preprocessing.minmax_scale(x, axis=axis)
librosa.display.waveshow(audio_file, sr=sr, alpha=0.4);
plt.plot(t, normalize(spectral_centroids), color='r');
plt.title("Spectrum centroid of flute", fontsize = 15);


# piano
plt.subplot(2,4,5)
y, sr = librosa.load(os.path.join(general_path,'piano.wav'))
audio_file, _ = librosa.effects.trim(y)
spectral_centroids = librosa.feature.spectral_centroid(audio_file, sr=sr)[0]
frames = range(len(spectral_centroids))
t = librosa.frames_to_time(frames)
def normalize(x, axis=0):
    return sklearn.preprocessing.minmax_scale(x, axis=axis)
librosa.display.waveshow(audio_file, sr=sr, alpha=0.4);
plt.plot(t, normalize(spectral_centroids), color='r');
plt.title("Spectrum centroid of piano", fontsize = 15);


# tenor saxophone
plt.subplot(2,4,6)
y, sr = librosa.load(os.path.join(general_path,'saxophone.wav'))
audio_file, _ = librosa.effects.trim(y)
spectral_centroids = librosa.feature.spectral_centroid(audio_file, sr=sr)[0]
frames = range(len(spectral_centroids))
t = librosa.frames_to_time(frames)
def normalize(x, axis=0):
    return sklearn.preprocessing.minmax_scale(x, axis=axis)
librosa.display.waveshow(audio_file, sr=sr, alpha=0.4);
plt.plot(t, normalize(spectral_centroids), color='r');
plt.title("Spectrum centroid of tenor saxophone", fontsize = 15);


# trumpet
plt.subplot(2,4,7)
y, sr = librosa.load(os.path.join(general_path,'trumpet.wav'))
audio_file, _ = librosa.effects.trim(y)
spectral_centroids = librosa.feature.spectral_centroid(audio_file, sr=sr)[0]
frames = range(len(spectral_centroids))
t = librosa.frames_to_time(frames)
def normalize(x, axis=0):
    return sklearn.preprocessing.minmax_scale(x, axis=axis)
librosa.display.waveshow(audio_file, sr=sr, alpha=0.4);
plt.plot(t, normalize(spectral_centroids), color='r');
plt.title("Spectrum centroid of trumpet", fontsize = 15);


# violin
plt.subplot(2,4,8)
y, sr = librosa.load(os.path.join(general_path,'violin.wav'))
audio_file, _ = librosa.effects.trim(y)
spectral_centroids = librosa.feature.spectral_centroid(audio_file, sr=sr)[0]
frames = range(len(spectral_centroids))
t = librosa.frames_to_time(frames)
def normalize(x, axis=0):
    return sklearn.preprocessing.minmax_scale(x, axis=axis)
librosa.display.waveshow(audio_file, sr=sr, alpha=0.4);
plt.plot(t, normalize(spectral_centroids), color='r');
plt.title("Spectrum centroid of violin", fontsize = 15);
plt.tight_layout()
plt.savefig("Spectrum centroid of 8 instruments.jpg")
```
